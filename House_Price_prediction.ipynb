```python
#Task 1: Importing the data
#--- Import Pandas ---
import pandas as pd
#--- Read in dataset ----
file_path= r'C:\Users\003UWG744\Downloads\HousePricePrediction.csv'
df = pd.read_csv(file_path)

#--- Inspect data ---
df
```

```python
# --- WRITE YOUR CODE FOR MODULE 2 TASK 1 ---
sumofnull = df.isnull().sum()
#--- Inspect data ---
print(sumofnull)
```

```python
# --- WRITE YOUR CODE FOR MODULE 2 TASK 2 ---
Smean=df['SalePrice'].mean()
#df = ...
df['SalePrice'].fillna(Smean,inplace=True)
```

```python
df1 = df.dropna()
df1
```

```python
data=df1.drop(columns='Id')
data
```

```python
#--- Import matplotlib and  seaborn---
import matplotlib.pyplot as plt
import seaborn as sns
corr = data.select_dtypes(include='number').corr()
print(corr)
f, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(corr, ax=ax, annot=True, square=True, cmap='coolwarm', fmt='.2f')
ax.set_title('Correlation Matrix Heatmap', pad=20)
plt.show()
```

```python
# --- WRITE YOUR CODE TASK 2 ---
cat_cols = ['MSZoning','LotConfig','BldgType','Exterior1st']
num_cols =['LotArea','TotalBsmtSF','MSSubClass','YearRemodAdd','BsmtFinSF2','YearBuilt','OverallCond','SalePrice']
counts={}
index=1
fig, axes = plt.subplots(len(cat_cols), 1, figsize=(12, 12))
for col in cat_cols:
    counts[col]=data[col].value_counts().to_dict()
    ax=axes[index-1]
    ct_sr=pd.Series(counts[col])
    ct_sr.plot(kind='bar',ax=ax)
    ax.set_title(col)
    ax.set_xlabel(col)
    ax.set_ylabel('Count')
    ax.tick_params(axis='x',rotation=90)
    index+=1
plt.tight_layout()
plt.subplots_adjust(top=0.9)
plt.show()
```

```python
df_encoded=pd.get_dummies(data,columns=cat_cols,prefix_sep='_')
print(df_encoded)
```

```python
#--- Import train_test_split from sklearn.model_selection ---
from sklearn.model_selection import train_test_split
# --- WRITE YOUR CODE FOR MODULE 4 TASK 2 ---
Y=df_encoded['SalePrice']
X=df_encoded.drop(columns=['SalePrice'])
X_train,X_valid,Y_train,Y_valid=train_test_split(X,Y,test_size=0.2,random_state=0)
#X_train,X_valid,Y_train, Y_valid = ...
#print(Y_train)
```

```python
#--- Import LinearRegression,RandomForestRegressor,svm,Ridge, mean_absolute_percentage_error,from sklearn ---
from sklearn.linear_model import LinearRegression,Ridge 
from sklearn.ensemble import RandomForestRegressor  
from sklearn.svm import SVR 
from sklearn.metrics import mean_absolute_percentage_error
#Intilize and fit Linear Regression model and find mape for lr model
model_LR=LinearRegression()
model_LR.fit(X_train,Y_train)
Y_pred=model_LR.predict(X_valid)
lr_mape = mean_absolute_percentage_error(Y_valid,Y_pred)
print(lr_mape)
#Intilize  and fit Random Forest model and find mape for rf model
model_RF=RandomForestRegressor()
model_RF.fit(X_train,Y_train)
Y_pred1=model_RF.predict(X_valid)
rf_mape = mean_absolute_percentage_error(Y_valid,Y_pred1)
print(rf_mape)

#Intilize  and fit svm model and find mape for svr model
#X_train2.replace({False:0,True:1},inplace=True)
model_SVR=SVR()
model_SVR.fit(X_train, Y_train)
Y_pred2=model_SVR.predict(X_valid)
svr_mape =mean_absolute_percentage_error(Y_valid,Y_pred2)
print(svr_mape)

#Intilize  and fit Ridge Regression model and find mape for ridge model
model_ridge=Ridge()
model_ridge.fit(X_train,Y_train)
Y_pred3=model_ridge.predict(X_valid)
ridge_mape =mean_absolute_percentage_error(Y_valid,Y_pred3)
print(ridge_mape)
```

```python
#--- Import  r2_score, mean_squared_error, mean_absolute_error from sklearn.metrics ---
#Choose the best model from previous task and find its other metrics
from sklearn.metrics import r2_score, mean_squared_error,mean_absolute_error
r2 = r2_score(Y_valid,Y_pred2)  # Calculate R-squared score
mae = mean_absolute_error(Y_valid,Y_pred2)  # Calculate Mean Absolute Error
rmse = mean_squared_error(Y_valid,Y_pred2,squared=False) # Calculate rmse
print(r2 ,'\n',mae, '\n', rmse ) 
```
